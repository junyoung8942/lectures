```python
# 피벗 테이블 : 개별 데이터 항목들으 집계 및 테이블 재구조화를 통해
#               데이터의 요약된(그룹화된) 결과를 나타내는 테이블
import pandas as pd
movies = pd.Series(['The Godfather, 1972, Francis Ford Coppola',
                   'Contact, 1997, Robert Zemeckis',
                   'Parasite, 2019, Joon-ho Bong'])
movies
```

```python
lst_title = []
lst_year = []
lst_director = []

for val in movies :
    title, year, director = val.split(',')
    lst_title.append(title)
    lst_year.append(year)
    lst_director.append(director)
print(lst_title)
print(lst_year)
print(lst_director)
```

```python
df_movie = pd.DataFrame()
df_movie['title'] = lst_title
df_movie['year'] = lst_year
df_movie['director'] = lst_director

df_movie
```

```python
import pandas as pd
import numpy as np
from pandas import datetime
from matplotlib import pyplot as plt
def parser(x) :
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
input_file = '../AirQualityUCI_refined.csv'

df = pd.read_csv(input_file,
                index_col=[0],
                parse_dates=[0],
                date_parser=parser)
df.head()
```

```python
# Visualization setup
%matplotlib
from matplotlib import pyplot as plt
import seaborn; seaborn.set()  # set plot styles
%config InlineBackend.figure_format = 'svg'
plt.rcParams['figure.figsize'] = [10, 5]
plt.ion() # enable the interactive mode

import seaborn as sns
sns.set()
```

```python
df.interpolate(inplace=True)
df.info()
```

```python
# Correlation Matrix
sns.pairplot(df, kind='reg', diag_kind='kde',
            plot_kws={'scatter_kws' : {'alpha' : 0.1}})
```

```python
# Prepare a feature set
# input features
x = df.iloc[:, 1:]
x
```

```python
# target variable
y = df.iloc[:, 0]
y
```

```python
# Feature Importance
from sklearn.ensemble import RandomForestRegressor

mode1 = RandomForestRegressor()
mode1.fit(x, y)

# Create and train model for regression
```

```python
# Plot the feature importances
print(mode1.feature_importances_)
```

```python
feat_importances = pd.Series(mode1.feature_importances_, index=x.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```
